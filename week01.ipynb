## 1.2 画像を表示する
### 真っ黒の画像を作成（大きさ $100\times 100$）

```python
img = np.zeros((100, 100, 3), dtype=np.uint8)
```

### 横線を引く

```python
img[50,:,0]=255 # R
img[50,:,1]=0   # G
img[50,:,2]=255 # B
```

### 図を表示

```python
plt.imshow(img)
```
